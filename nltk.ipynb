```python
d0 = """The simplest example is provided by deterministic nodes. A deterministic node has "ArithmeticError"""
d1 = """The relationship can also be numerical: for example, if the parent nodes are the parent"""
d2 = """In fact, this is a worst-case scenario in which the relationship between the parent"""
```

```python
import pandas as pd
string = [d0,d1,d2]
tfidf = TfidfVectorizer(ngram_range = (1,1))
result = tfidf.fit_transform(string)
feature_names=tfidf.get_feature_names_out()
dense=result.todense()
dens_lst=dense.tolist()
df=pd.DataFrame(dens_lst,columns=feature_names)
```
